```python
# Make sure column renaming is already done on deathtime_0331
deathttime_0331 = deathttime_0331.rename(columns={
    "deathtime": "deathtime_1",
    "hospital_expire_flag": "expire_flag_1"
})

# Merge only on matching hadm_id values
merged_df = ak_df_cleaned.merge(deathttime_0331, on="hadm_id", how="inner")

```

```python
# Step 1: Drop the original expire_flag and DOD columns
merged_df = merged_df.drop(columns=["expire_flag", "DOD"])

# Step 2: Rename the new columns
merged_df = merged_df.rename(columns={
    "expire_flag_1": "expire_flag",
    "deathtime_1": "DOD"
})

merged_df.head()
```

```python
pd.set_option('display.max_rows', 237)

# Now display the subset
merged_df[['DOD', 'ADMITTIME']]
```

```python
merged_df["DOD_str"] = merged_df["DOD"].fillna("").astype(str)
```

```python
year_diffs = []
month_diffs = []

for idx, row in merged_df.iterrows():
    dod = row["DOD"]
    admit = row["ADMITTIME"]

    if pd.notna(dod):
        dod_str = str(dod)
        admit_str = str(admit)

        # Extract year and month
        dod_year = int(dod_str[:4])
        dod_month = int(dod_str[5:7])
        admit_year = int(admit_str[:4])
        admit_month = int(admit_str[5:7])

        # Calculate diffs
        year_diff = dod_year - admit_year
        month_diff = dod_month - admit_month

        year_diffs.append(year_diff)
        month_diffs.append(month_diff)
    else:
        year_diffs.append(np.nan)
        month_diffs.append(np.nan)
```

```python
merged_df["year_diff"] = year_diffs
merged_df["month_diff"] = month_diffs
```

```python
ak_cleaned = merged_df
```

```python
ak_cleaned = ak_cleaned.drop(columns=["Year diff.", "Month diff.", 'month_diff_match', 'year_diff_match'])

# Step 2: Rename the new ones
ak_cleaned = ak_cleaned.rename(columns={
    "year_diff": "Year diff.",
    "month_diff": "Month diff."
})
```

```python
ak_cleaned = ak_cleaned.drop(columns=["DISCHTIME"], errors="ignore")

# Step 2: Merge on 'hadm_id', keeping only rows from ak_cleaned
experimental_df = ak_cleaned.merge(discharges, on="hadm_id", how="left")

# Step 3: Rename 'dischtime' to 'DISCHTIME'
experimental_df = experimental_df.rename(columns={"dischtime": "DISCHTIME"})

experimental_df.head()
```

```python
experimental_df = experimental_df.rename(columns={"DOD_str": "DOD"})
```

```python
experimental_df.columns
```

```python
not_ak_df = v8_asah[v8_asah["Extractor"] != "AK"]
```

```python
# Step 1: Get column differences
experimental_cols = set(experimental_df.columns)
not_ak_cols = set(not_ak_df.columns)

if experimental_cols != not_ak_cols:
    missing_in_not_ak = experimental_cols - not_ak_cols
    missing_in_experimental = not_ak_cols - experimental_cols

    raise ValueError(f"Column mismatch!\n"
                     f"→ Columns in experimental_df but not in not_ak_df: {missing_in_not_ak}\n"
                     f"→ Columns in not_ak_df but not in experimental_df: {missing_in_experimental}")
else:
    # Step 2: If columns match, combine the dataframes
    final_df = pd.concat([experimental_df, not_ak_df], ignore_index=True)
    print("DataFrames successfully combined.")
```
