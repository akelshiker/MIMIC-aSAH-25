```python
import numpy as np
import pandas as pd
```

```python
v2 = pd.read_csv('/content/drive/MyDrive/Assisting JK AG Paper/v2 aSAH ASM 2024.csv')

new_1113 = pd.read_csv('/content/drive/MyDrive/12_12_24_datarevisiting/v_2new_1113_toread.csv')
```

```python
genders_1113 = pd.read_csv('/content/drive/MyDrive/12_12_24_datarevisiting/genders_1113.csv')
```

```python
merged_new_1113 = pd.merge(new_1113, genders_1113, on='subject_id', how='inner')
```

```python
# adjusting ADMITTIME DISCHTIME
# Ensure 'ASM_start' and 'ASM_end' are in datetime format
merged_new_1113['ASM_start'] = pd.to_datetime(merged_new_1113['ASM_start']).dt.date
merged_new_1113['ASM_end'] = pd.to_datetime(merged_new_1113['ASM_end']).dt.date
merged_new_1113.head()
```

```python
merged_new_1113['mingcs'] = merged_new_1113['minGCS (from ICUdata)']
```

```python
merged_new_1113 = merged_new_1113.drop(columns=['anchor_year_group'], errors='ignore')
```

```python
# Rename columns
merged_new_1113 = merged_new_1113.rename(columns={
    'los_hospital': 'los_icu',
    'los_hospital.1': 'los_hospital',
    'gender' : "GENDER",
    'rrt' : "RRT"
})
```

```python
# Columns unique to merged_new_1113
columns_in_merged_not_in_v2 = set(merged_new_1113.columns) - set(v2.columns)

# Columns unique to v2
columns_in_v2_not_in_merged = set(v2.columns) - set(merged_new_1113.columns)

# Print results
print("Columns in merged_new_1113 but not in v2:", columns_in_merged_not_in_v2)
print("Columns in v2 but not in merged_new_1113:", columns_in_v2_not_in_merged)

# Columns common to both
common_columns = set(merged_new_1113.columns).intersection(set(v2.columns))
print("Columns common to both:", common_columns)
```

```python
# Step 1: Ensure both DataFrames have the same columns
merged_new_1113 = merged_new_1113[v2.columns]  # Reorder columns to match v2

# Step 2: Concatenate the DataFrames
combined_df = pd.concat([v2, merged_new_1113], ignore_index=True)
```

```python
genders_1416 = pd.read_csv('/content/drive/MyDrive/12_12_24_datarevisiting/genders_1416.csv')
genders_1416.head()
```

```python
new_1416 = pd.read_csv('/content/drive/MyDrive/12_12_24_datarevisiting/new_14_16_toread.csv')
```

```python
merged_1416 = pd.merge(new_1416, genders_1416, on='subject_id', how='inner')
```

```python
# Step 1: Rename columns
merged_1416 = merged_1416.rename(columns={
    'gender': 'GENDER',
    'ASM_stop': 'ASM_end'
})

# Step 2: Remove 'anchor_year_group' column if it exists
if 'anchor_year_group' in merged_1416.columns:
    merged_1416 = merged_1416.drop(columns=['anchor_year_group'])

# Step 3: Convert ADMITTIME and DISCHTIME to date-only format
merged_1416['ADMITTIME'] = pd.to_datetime(merged_1416['ADMITTIME']).dt.date
merged_1416['DISCHTIME'] = pd.to_datetime(merged_1416['DISCHTIME']).dt.date

```

```python
# Get the column names as sets
merged_columns = set(merged_1416.columns)
combined_columns = set(combined_df.columns)

# Find differences
columns_in_merged_not_in_combined = merged_columns - combined_columns
columns_in_combined_not_in_merged = combined_columns - merged_columns

# Print the differences
print("Columns in merged_1416 but not in combined_df:", columns_in_merged_not_in_combined)
print("Columns in combined_df but not in merged_1416:", columns_in_combined_not_in_merged)

# Print common columns
common_columns = merged_columns.intersection(combined_columns)
print("Common columns:", common_columns)
```

```python
# Step 1: Align columns to match combined_df's order
merged_1416 = merged_1416[combined_df.columns.intersection(merged_1416.columns)]

# Step 2: Concatenate the DataFrames
df = pd.concat([combined_df, merged_1416], ignore_index=True)
df.head()

```
